Layer 1: Physical Layer

- Signalling, cabling, connectors
- This layer isn't about protocols

When referring to "You have a physical layer problem"
- We would try and fix cabling, punch-downs, etc...
- Run loopback tests, test/replace cables, swap adapter cards. 


Layer 2 : Data Link Layer

- The basic network "language"
- The foundation of communication at the data link layer

Data link Control (DLC) protocols
- MAC (Media Access Control) address on Ethernet


Layer 3: Network Layer
- The "routing" layer
- Associated with IP addresses
- Fragment frames to traverse different networks


Layer 4: Transport Layer
- The "post office" layer
    - Parcels and letters
- Uses TCP (Transmission Control Protocol) and UDP (User Datagram Protocol)
    Sends the website in different frames and the user will than arrange the 
    the frames back into order once received.


Layer 5 : Session Layer
- Communication management
    between devices (start, stop and restart)
- Control protocols, tunneling protocols (set up a session)


Layer 6: Presentation Layer (SSL/TLS)
- Character encoding
- Application Encryption
- Often combined with the application layer

Layer 7: Application Layer (Your EYES)
- This is the layer that we can see
HTTP, FTP, DNS, POP3 

e.g

Application: https://mail.google.com
Presentation: SSL encryption
Session: Link the presentation to the transport
Transport: TCP encapsulation
Network: IP encapsulation
Data Link: Ethernet
Physical: Electrical Signals



Data Communication:
How do we get data to move accross the network?
    We require something called PDU (Protocol Data Unit)

- Transmission Units:
    A different group of data at different OSI layers


Sending information from the source to the destination.
In order for the source to transfer the application data it must first encapsulate
the data by each layer's relevant PDUs.

e.g Source: Frame Header, IP Header, TCP Header, Application Data, Frame Trailer

The source encapsulates the Application data with all these headers, which the 
destination header will decapsulate to recieve the application data.

Each layer has a header and payload
